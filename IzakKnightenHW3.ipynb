```python
#Block 10
%%sql
DROP TABLE IF EXISTS Marketing_Brands_Asc;

CREATE TABLE Marketing_Brands_Asc AS
SELECT *
FROM Marketing_Brands
ORDER BY brand_budget ASC;
```
